```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd 
from tqdm import tqdm
```

```python
link = "https://quotes.toscrape.com/"
res = requests.get(link)
```

```python
html = res.text
```

```python
fd = open("main.html" ,"w",encoding="utf-8")
fd.write(html)
fd.close()
```

```python
soup = BeautifulSoup(html , "html.parser")
```

```python
print(soup.find("span", class_ = "text").text[1:-1])
```

```python
for i in soup.find_all("span", class_ = "text"):
    print(i.text[1:-1])
```

```python
sp = BeautifulSoup(html ,  "html.parser")
soup = sp.find_all("div" , class_ = "quote")
for i in soup:
    k = i.find("small")
    print(k.text ,end =":  ")
        
    j = i.find("span", class_ = "text")
    print(j.text )
    
    m = i.find("a")
    print(m.get("href"))
    print(m.text)
    print("\n")
```

```python
sp = BeautifulSoup(html ,  "html.parser")
soup = sp.find_all("div" , class_ = "quote")
for i in soup:
    for k in i.find_all("small"):
        print(k.text ,end =":  ")
        
    for j in  i.find_all("span", class_ = "text"):
        print(j.text )
    
    for m in  i.find_all("a"):
        print(m.get("href"))
        print(m.text)
    print("\n")
```

```python
sp = BeautifulSoup(html ,  "html.parser")
soup = sp.find_all("div" , class_ = "quote")
lst = []
for i in soup:
    k = i.find("small")
    author = k.text
        
    j = i.find("span", class_ = "text")
    quote = j.text[1:-1]
    
    
    m = i.find("a",class_ = 'tag')
    author_id = m.get("href")
    l = []
    for m in  i.find_all("a",class_ = 'tag'):
        l.append(m.text)
    l = ",".join(l)
    lst.append([author,quote,author_id,l])
    
        
    
```

```python
lst
```

```python
df = pd.DataFrame(lst,columns = ['author','quote','author_id','tags']) 
```

```python
df
```

```python
df.to_csv('quotes.csv')
```

## multiple webpages

```python
data = []
for page in tqdm(range(1,11)):
    
    link = "https://quotes.toscrape.com/page/"+str(page)+"/"
    res = requests.get(link)
    soup = BeautifulSoup(res.text, "html.parser")
    sp = soup.find_all("div" ,class_ = "quote")
    for i in sp:
        quotes = i.find("span",class_ = "text").text[1:-1]
        
        author = i.find("small").text
        
        m = i.find("a")
        author_id = m.get("href")
        ls = []
        for j in i.find_all("a" ,class_ = "tag"):
            ls.append(j.text)
        tags = ",".join(ls)
        
        data.append([author , quotes , author_id , tags])

        
        
    
```

```python
data
        
    
```

```python
df = pd.DataFrame(data,  columns = ['author','quotes','author_id','tags'])
```

```python
df['author_link'] = 'https://quotes.toscrape.com'+df['author_id']
```

```python
df.to_csv('multipleQuotes.csv')
```
