```python
# from selenium import webdriver
# import chromedriver_binary as cm
# from bs4 import BeautifulSoup
# cm.chromedriver_filename
# driver = webdriver.Chrome(cm.chromedriver_filename)
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup


# Automatically download the correct ChromeDriver
service = Service(ChromeDriverManager().install())
options = Options()

# Initialize the driver
driver = webdriver.Chrome(service=service, options=options)

```

```python
driver.get("https://www.youtube.com/c/GeeksForGeeksVideos/videos")
```

```python
soup = BeautifulSoup(driver.page_source , "html.parser")
```

```python
# len(soup.find_all("ytd-rich-item-renderer"))
sp = soup.find_all("ytd-rich-item-renderer")
```

```python
link = sp[0].find("a" ,class_ = "yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media").get("href")
```

```python
title = sp[0].find("a" ,class_ = "yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media").text
```

```python
views = sp[0].find_all("span" , class_ = "inline-metadata-item style-scope ytd-video-meta-block")[0].text
```

```python
date_time = sp[0].find_all("span" , class_ = "inline-metadata-item style-scope ytd-video-meta-block")[1].text
```

```python
image_link = sp[1].find("a" , class_ = "yt-simple-endpoint inline-block style-scope ytd-thumbnail").find("img").get("src")
```

```python
import pandas as pd
```

```python
data = []
for i in sp:
    link = i.find("a" ,class_ = "yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media").get("href")
    title = i.find("a" ,class_ = "yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media").text
    try:
        
        views = i.find_all("span" , class_ = "inline-metadata-item style-scope ytd-video-meta-block")[0].text
    except:
        views = None
    try:
        date_time = i.find_all("span" , class_ = "inline-metadata-item style-scope ytd-video-meta-block")[1].text
    except:
        date_time = None
    try:
        image_link = i.find("a" , class_ = "yt-simple-endpoint inline-block style-scope ytd-thumbnail").find("img").get("src").split("?")[0]
    except:
        image_link = None
    data.append([title , link , views , date_time , image_link ])
    
```

```python
df  = pd.DataFrame(data , columns = ["title" , "link" , "views" , "date_time" , "image_link" ])
```
